```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customer Churn.csv')
df.head()
```

```python
df.info()
```

# Replacing blanks with 0 and no total charges are recorded

# Datasheet change to Table formatting

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)


```

```python
df.head(30)
```

# converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn ")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customer", fontsize = 10)
plt.show()
```

# From the given pie chart we can conclude that 26.54% of our customers have out.
# Not let's explore the reson behind it.

```python
plt.figure(figsize = (5,5))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (5,5))
sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
plt.title("Churn by SeniorCitizen")
plt.show()
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack()

# Plot
fig, ax = plt.subplots(figsize=(6, 6))

# plot the bars
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])

# Add percentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

plt.title('Churn by SeniorCitizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', loc='upper right')

plt.show()
```

```python
categorical_cols = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents',
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn'
]

# Step 2: Set up subplot grid
n_cols = 3  # number of columns in subplot layout
n_rows = int(np.ceil(len(categorical_cols) / n_cols))

plt.figure(figsize=(n_cols * 6, n_rows * 4))

# Step 3: Loop through columns and plot countplots
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.countplot(x=col, data=df, hue='Churn')
    plt.title(f'Churn by {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.show()
```

# Senior citizens and those with no dependents or partners have a noticeably higher churn rate.Customers without online security, backup, or tech support also show increased churn.Internet service type impacts churn—those with fiber optic are more likely to leave compared to DSL or no service.Features like PhoneService and MultipleLines have a visible difference but less drastic churn separation.
